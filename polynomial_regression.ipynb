```python
import pandas as pd
```

```python
s1 = pd.Series(X ,name="Time")
```

```python
s1
```

```python
s2 = pd.Series(y ,name="speed")
```

```python
s2
```

```python
df = pd.DataFrame((s1,s2))
df
```

```python
df = pd.concat([s1,s2],axis=1)
df
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf = PolynomialFeatures()
```

```python
x1 = pf.fit_transform(df[["Time"]])
x1
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x1,df.speed)
```

```python
coef = model.coef_
coef
```

```python
interept = model.intercept_
interept
```

```python
y_predicted = interept + line * coef[1] + coef[2]*line**2
y_predicted
```

```python
plt.scatter(df['Time'],df['speed'])
plt.plot(y_predicted,color = 'r')
plt.show()
```

```python
model.score(x1,df['speed'])
```
