```python
from sklearn.datasets import load_iris
```

```python
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'] , label = "Setosa")
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],label = "versicolor")
plt.scatter(df2['sepal length (cm)'],df2['sepal width (cm)'],label = "verginica")
plt.xlabel("sepal length")
plt.ylabel("sepal width")
plt.title("sepal distribution")
plt.legend()
plt.show()
```

```python
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'], label = "Setosa")
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'], label = "verisa")
plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'], label = "vercanica")
plt.legend()
plt.show()
```

```python
X = df.drop("target",axis=1)
y = df.target
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20 ,random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.predict([[4.5,3.5,1.5,0.4]])
```

```python
score = []
for i in range(1,15):
    model=KNeighborsClassifier(n_neighbors=i)
    model.fit(X_train,y_train)
    accuracy = model.score(X_test,y_test)
    score.append(accuracy)
    
```

```python
score
```

```python
plt.plot(range(1,15),score)
plt.show()
```

```python
y_pred = model.predict(X_test)
```

```python
cm = confusion_matrix(y_test,y_pred)
```

```python
import seaborn as sns
```

```python
sns.heatmap(cm,annot=True)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```
